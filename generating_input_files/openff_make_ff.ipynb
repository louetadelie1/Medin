```python
from openff.units import unit
import numpy as np
from openff.interchange import Interchange
from openff.toolkit import ForceField
from openff.toolkit import Molecule, Topology
```

```python
forcefield = ForceField("openff-2.2.0.offxml")
sage = ForceField("openff-2.0.0.offxml")
cubic_box = unit.Quantity(30 * np.eye(3), unit.angstrom)
```

#### Molecule CM8 - ZINC000265540661

```python
sdf_cm8 = Molecule.from_file("/Users/adelielouet/Documents/science/medin/cm8/ZINC000265540661.sdf")
sdf_cm8.visualize()

```

```python

sdf_cm8.generate_conformers()
sage = ForceField("openff-2.0.0.offxml")
cubic_box = unit.Quantity(30 * np.eye(3), unit.angstrom)

interchange_cm8 = Interchange.from_smirnoff(topology=[sdf_cm8], force_field=sage, box=cubic_box)

#interchange_cm8.to_gromacs("/Users/adelielouet/Documents/science/medin/cm8/parameter_files_openff/cm8")  # Produces the same three files
```

```python
interchange_cm8.to_gromacs("/Users/adelielouet/Documents/science/medin/cm10/parameter_files_openff/cm8",monolithic=False)  
```
